```python
# Import the pandas library
# Pandas is a popular library for data manipulation and analysis
import pandas as pd

# Step 1: Define the data
# Create a dictionary where each key represents a column name
# Each key maps to a list of values (data) for that column
# NOTE: The lists must have the same length, as each element in the list represents a row in the table
#       For example, the first element in each list represents the first row in the table 1500, 2, 20, 1.5, 250
data = {
    "Square Footage": [1500, 2000, 1200, 1800, 2500],  # List of square footage values
    "Number of Bathrooms": [2, 3, 1, 2, 4],           # List of the number of bathrooms
    "House Age (Years)": [20, 10, 50, 15, 5],         # List of house ages in years
    "Distance to School (Miles)": [1.5, 2.0, 0.8, 1.2, 3.0],  # Distance to school in miles
    "House Price (Label)": [250, 400, 150, 300, 500]  # House prices (labels)
}

# Optional alternative way to create the dictionary:
# data = [
#     {
#         "Square Footage": 1500,
#         "Number of Bathrooms": 2,
#         "House Age (Years)": 20,
#         "Distance to School (Miles)": 1.5,
#         "House Price (Label)": 250
#     },
#     {
#         "Square Footage": 2000,
#         "Number of Bathrooms": 3,
#         "House Age (Years)": 10,
#         "Distance to School (Miles)": 2.0,
#         "House Price (Label)": 400
#     },
#     {
#         "Square Footage": 1200,
#         "Number of Bathrooms": 1,
#         "House Age (Years)": 50,
#         "Distance to School (Miles)": 0.8,
#         "House Price (Label)": 150
#     },
#     {
#         "Square Footage": 1800,
#         "Number of Bathrooms": 2,
#         "House Age (Years)": 15,
#         "Distance to School (Miles)": 1.2,
#         "House Price (Label)": 300
#     },
#     {
#         "Square Footage": 2500,
#         "Number of Bathrooms": 4,
#         "House Age (Years)": 5,
#         "Distance to School (Miles)": 3.0,
#         "House Price (Label)": 500
#     }
# ]


# Step 2: Create a DataFrame
# A DataFrame is a table-like structure provided by pandas
# Pass the dictionary to the pandas DataFrame constructor
df = pd.DataFrame(data)

# Step 3: Adjust the Display
# Adjust pandas options to display more columns and increase the output width
pd.set_option("display.width", 1000)  # Set the desired width (1000 is an example)
pd.set_option("display.max_columns", None)  # Show all columns without truncation

# Step 4: Display the DataFrame
# Print the DataFrame to the console to see its content
print(df)

# Explanation of output:
# The output will be a table where:
# - Each key in the dictionary becomes a column header
# - Each list in the dictionary becomes the rows under that column
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Original dataset
data = {
    "Square Footage": [1500, 2000, 1200, 1800, 2500],
    "House Price (Label)": [250, 400, 150, 300, 500]
}

# Create the DataFrame
df = pd.DataFrame(data)

# Plot the data using Matplotlib
plt.figure(figsize=(8, 5))

# Scatter plot for Square Footage vs. House Price
plt.scatter(df["Square Footage"], df["House Price (Label)"], 
            color="blue", label="House Price vs. Square Footage", alpha=0.7)

# Add labels, title, and legend
plt.title("House Price vs. Square Footage (Matplotlib)", fontsize=14)
plt.xlabel("Square Footage", fontsize=12)
plt.ylabel("House Price (Label)", fontsize=12)
plt.legend()
plt.grid(alpha=0.3)

# Show the plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Original dataset
data = {
    "Square Footage": [1500, 2000, 1200, 1800, 2500],
    "House Price (Label)": [250, 400, 150, 300, 500]
}

# Create the DataFrame
df = pd.DataFrame(data)

# Set the Seaborn style
sns.set_theme(style="darkgrid", palette="deep")

# Create a scatter plot with a regression line
plt.figure(figsize=(10, 6))
sns.regplot(
    x="Square Footage",
    y="House Price (Label)",
    data=df,
    scatter_kws={"color": "blue", "s": 80, "alpha": 0.8},  # Scatter points style
    line_kws={"color": "red", "linewidth": 2},  # Regression line style
    ci=95  # Confidence interval for the regression line
)

# Add title and labels
plt.title("House Price vs. Square Footage (Seaborn - Enhanced)", fontsize=16, fontweight="bold")
plt.xlabel("Square Footage", fontsize=14)
plt.ylabel("House Price (Label)", fontsize=14)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Original dataset
data = {
    "Square Footage": [1500, 2000, 1200, 1800, 2500, 
                        1200, 3000, 2500, 1800, 4000],
    "Number of Bathrooms": [2, 3, 1, 2, 4, 
                            3, 2, 1, 5, 0],
    "House Age (Years)": [20, 10, 50, 15, 5, 
                            60, 3, 45, 100, 1],
    "Distance to School (Miles)": [1.5, 2.0, 0.8, 1.2, 3.0, 
                                    1.1, 2.5, 0.0, 0.9, 0.6],
    "House Price (Label)": [250, 400, 150, 300, 500, 
                            220, 750, 90, 350, 1000]
}

# Create the DataFrame
df = pd.DataFrame(data)

# Remove outliers
df_filtered = df[
    (df["House Price (Label)"] != 750) &
    (df["House Price (Label)"] != 1000) &
    (df["Distance to School (Miles)"] != 0)
]

# Check the filtered data
print("Filtered DataFrame:")
print(df_filtered)

# Scatter plot with regression line
plt.figure(figsize=(8, 6))
sns.regplot(x="Distance to School (Miles)", y="House Price (Label)", data=df_filtered,
            scatter_kws={"color": "blue", "s": 100, "alpha": 0.8},
            line_kws={"color": "red", "linewidth": 2})

# Add title and labels
plt.title("Scatter Plot with Regression Line (Filtered Data)", fontsize=16)
plt.xlabel("Distance to School (Miles)", fontsize=14)
plt.ylabel("House Price (Label)", fontsize=14)

# Show the plot
plt.tight_layout()
plt.show()

```
